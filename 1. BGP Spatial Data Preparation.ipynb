**DATA CLEANING AND FILTERING**

```python
# EXPORTING BLOCK(SQL) TO CSV

block_results = %sql \
select b.block, b.longit, b.latit, b.geom  \
from block as b \
where b.latit != 0 and b.latit!= 393 and b.block <= 43000 \
order by b.block asc \

block = block_results.DataFrame()
block.to_csv("block.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING OBSERVER INTERNSITY(SQL) TO CSV

observer_intensity_results = %sql \
select o.block, COUNT(distinct o.observer) AS observer_intensity \
from observation as o \
where o.block <= 43000 \
group by o.block \
order by o.block asc \

observer_intensity = observer_intensity_results.DataFrame()
observer_intensity.to_csv("observer intensity.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING BIRD-BIODIVERSITY(SQL) TO CSV

birds_results = %sql \
select bio.block, bio.aves2012_2016 as birds \
from biodiversity as bio \
where bio.block <= 43000 \
order by bio.block asc \

birds = birds_results.DataFrame()
birds.to_csv("birds.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING SURROUNDING DEMOGRAPHY(SQL) TO CSV

surr_demography_results = %sql \
select d.block, d.pop225 as surr_pop \
from block_demography as d \
where d.block <= 43000 \
order by d.block asc \

surr_demography_results = surr_demography_results.DataFrame()
surr_demography_results.to_csv("surrounding demography.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING ABSOLUTE DEMOGRAPHY(SQL) TO CSV

absolute_demography_results = %sql \
select d.block, d.pop001 as abs_pop \
from block_demography as d \
where d.block <= 43000 \
order by d.block asc \

absolute_demography_results = absolute_demography_results.DataFrame()
absolute_demography_results.to_csv("absolute demography.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING NATURAL RESERVES LAND COVERAGE(SQL) TO CSV

natural_land_results = %sql \
select l.block, l.areasum as nat_cover \
from block_landuse as l \
where l.category = 'nature reserves' and l.block <= 43000 \
order by l.block asc \

natural_land = natural_land_results.DataFrame()
natural_land.to_csv("natural_land.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING BUILT-UP LAND COVERAGE(SQL) TO CSV

builtup_land_results = %sql \
select l.block, l.areasum as built_cover \
from block_landuse as l \
where l.category = 'Bebouwd' and l.block <= 43000 \
order by l.block asc \

builtup_land = builtup_land_results.DataFrame()
builtup_land.to_csv("builtup_land.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING ROAD ACCESS COVERAGE TOTAL (SQL) TO CSV

road_access_results = %sql \
select r.block, sum(r.roadlength) as roadlength \
from block_road_access as r \
where r.block <= 43000 \
group by r.block \
order by r.block asc \

road_access = road_access_results.DataFrame()
road_access.to_csv("road access.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING ROAD ACCESS COVERAGE MUNICIPAL (SQL) TO CSV

road_access_municipal = %sql \
select r.block, sum(r.roadlength) as G_roadlength \
from block_road_access as r \
where r.block <= 43000 and r.maintainer = 'G'\
group by r.block \
order by r.block asc \

road_access_municipal = road_access_municipal.DataFrame()
road_access_municipal.to_csv("road access municipal.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING ROAD ACCESS COVERAGE PROVINCE (SQL) TO CSV

road_access_province = %sql \
select r.block, sum(r.roadlength) as P_roadlength \
from block_road_access as r \
where r.block <= 43000 and r.maintainer = 'P'\
group by r.block \
order by r.block asc \

road_access_province = road_access_province.DataFrame()
road_access_province.to_csv("road access province.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING AVERAGE PRECIPITATION(SQL) TO CSV

precipitation_results = %sql \
select p.block, avg(p.precip) AS ave_precip \
from precipitation as p \
where p.block <= 43000 \
group by p.block \
order by p.block asc \

precipitation = precipitation_results.DataFrame()
precipitation.to_csv("precipitation.csv",
             index = False,
             encoding = 'utf-8-sig')
```

```python
# EXPORTING AVERAGE TEMPERATURE(SQL) TO CSV

temperature_results = %sql \
select t.block, avg(t.temper) AS ave_temper \
from temperature as t \
where t.block <= 43000 \
group by t.block \
order by t.block asc \

temperature = temperature_results.DataFrame()
temperature.to_csv("temperature.csv",
             index = False,
             encoding = 'utf-8-sig')
```

**JOINING MULTIPLE TABLES WITH 'BLOCK' AS PRIMARY KEY**

```python
# loading multiple dataset
filenames = ['block.csv',
             'birds.csv',
             'natural_land.csv',
             'builtup_land.csv',
             'observer intensity.csv',
             'precipitation.csv',
             'road access.csv',
             'road access province.csv',
             'road access municipal.csv',
             'surrounding demography.csv',
             'absolute demography.csv',
             'temperature.csv']

# concatenate and combine with block as primary key
files_combined = [pd.read_csv(filename, index_col = 'block' ) for filename in filenames]
all_combined_data = files_combined[0].join(files_combined[1:])

all_combined_data
```

```python
# removing null values
clean_data = all_combined_data.dropna()

clean_data
```

**EXPORTING FINAL DATA**

```python
# export to csv
all_combined_data.reset_index(inplace = True)
all_combined_data.to_csv("all spatial data.csv",
                    index = False,
                    encoding = 'utf-8-sig')

clean_data.reset_index(inplace = True)
clean_data.to_csv("spatial data without null values.csv",
                    index = False,
                    encoding = 'utf-8-sig')
```
